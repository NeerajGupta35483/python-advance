**Q8. What is the benefit of first checking the target string with in or
not in before using the index method to find a substring ?**

**Ans:** Checking the target string with **in** or **not** Operators
before using the index method to find a substring just helps confirming
availability of substring and thus avoid raising of **ValueError**.

Example: in_string = "ineuron" in_string.index('x') \# Raises ValueError
in_string.index('u') \# 3
