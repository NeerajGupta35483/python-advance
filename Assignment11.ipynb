Q1. What is the concept of a metaclass?**

**Ans:** Metaclass in Python is a class of a class that defines how a
class behaves. A class is itself a instance of Metaclass, and any
Instance of Class in Python is an Instance of type metaclass. E.g. Type
of of int, str, float, list, tuple and many more is of metaclass type.

**Q2. What is the best way to declare a class's metaclass?**
