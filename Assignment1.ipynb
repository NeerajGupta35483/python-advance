Q1. What is the purpose of Python's OOP?**

**Ans:** Object-oriented programming is a programming paradigm that
provides a means of structuring programs so that properties and
behaviors are bundled into individual objects

-   In Python, object-oriented Programming (OOPs) uses objects and
    classes in programming.

-   It aims to implement real-world entities like inheritance,
    polymorphisms, encapsulation, etc. in the programming.

-   The main concept of OOPs is to bind the data and the functions that
    work on that together as a single unit so that no other part of the
    code can access this data.

-   It comes up with the following advantages:

-   It helps to divide our over all program into different small
    segments and thus making it solving easy with the use of objects

-   Helps in easy maintenance and modification of existing program

-   Multiple instances of an object can be made.

-   Class Methods (class level methods)
